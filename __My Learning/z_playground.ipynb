# Drawing a Poisson Distribution Chart

```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters
lambda_ = 7.5  # Average rate (mean number of events)
size = 1000  # Number of samples
resolution = 25 # x-axis scale (number of events)

# Generate Poisson-distributed data
data = np.random.poisson(lambda_, size)

# Plot the distribution
sns.histplot(data, kde=False, stat='density', bins=range(0, resolution))
plt.title('Poisson Distribution (λ=3)')
plt.xlabel('Number of events')
plt.ylabel('Probability')
plt.show()
```
