# Drawing a Poisson Distribution Chart

```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters
lambda_ = 100  # Average rate (mean number of events)
size = 10000  # Number of samples
resolution1 = 60 # x-axis scale (min number of events)
resolution2 = 150 # x-axis scale (max number of events)

# Generate Poisson-distributed data
data = np.random.poisson(lambda_, size)

# Plot the distribution
sns.histplot(data, kde=False, stat='density', bins=range(resolution1, resolution2))
plt.title('Poisson Distribution (Î»=3)')
plt.xlabel('Number of events')
plt.ylabel('Probability')
plt.show()
```
